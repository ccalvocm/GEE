```python
import os
import ee
import geemap
```

Upgrade geemap to the latest version

```python
geemap.update_package()
```

```python
Map = geemap.Map()
Map
```

```python
landsat7 = ee.ImageCollection('UCSB-CHG/CHIRPS/DAILY').filter(ee.Filter.date('2023-08-22', '2024-08-24')).first()
landsat7 = ee.ImageCollection('NASA/GPM_L3/IMERG_V06').filter(ee.Filter.date('2023-08-22', '2024-08-24')).first()

# Set visualization parameters.
vis_params = {
    'min': 0,
    'max': 400,
    'palette': ['006633', 'E5FFCC', '662A00', 'D8D8D8', 'F5F5F5'],
}

# Add Earth Engine layers to Map
Map.addLayer(
   landsat7, {'bands': 'precipitation', 'min': 0, 'max': 200}, 'Pp'
)
```

Download sample data

```python
work_dir='.'
in_shp = os.path.join(work_dir, 'Maipo.shp')
in_fc = geemap.shp_to_ee(in_shp)
print(in_fc)
Map.addLayer(in_fc, {}, 'cuencaMaipo')
```

Export pixel values as a csv

```python
out_csv = os.path.join(work_dir, 'chirps.csv')
geemap.extract_values_to_points(in_fc, landsat7, out_csv,scale=5000, projection='EPSG:4326', tileScale=1, getInfo=True)
import pandas as pd
print(pd.read_csv(out_csv))
```
